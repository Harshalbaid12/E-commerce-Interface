```python
import matplotlib.pyplot as plt
import pandas as pd
from google.colab import files
uploaded = files.upload()
```

```python
stress = pd.read_excel("Dataset.xlsx")
stress
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
x = stress.iloc[:,:-1]
y = stress.Stress

x = x.apply(encoder.fit_transform)
x
```

```python
from sklearn.tree import DecisionTreeClassifier
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,confusion_matrix,ConfusionMatrixDisplay,classification_report
from sklearn.neighbors import KNeighborsClassifier

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=1/4,random_state=1)
regressor = DecisionTreeClassifier()
decision_tree = regressor.fit(x_train,y_train)
y_pred_dt = decision_tree.predict(x_test)
score = accuracy_score(y_test,y_pred_dt) 
print(score)


```

KNN Classifier


```python
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors=10,metric='manhattan')
classifier.fit(x_train, y_train)
y_pred_knn = classifier.predict(x_test)
score = accuracy_score(y_test,y_pred_knn)
print(score)
```

```python
print(confusion_matrix(y_test, y_pred_dt))
print(confusion_matrix(y_test, y_pred_knn))
```

```python
print(classification_report(y_test, y_pred_dt))
print(classification_report(y_test, y_pred_knn))
```
